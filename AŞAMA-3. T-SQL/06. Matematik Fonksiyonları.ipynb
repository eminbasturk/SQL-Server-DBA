* **COS**
* **COT**
* **DEGREES**
* **EXP**
* **FLOOR**: Bir ondalık değeri aşağı yuvarlar.
* **LOG**
* **LOG10**
* **PI**
* **POWER**: Herhangi bir sayının üssünü alır.
* **RADIANS**
* **RAND**: Rastgele değer üretir.
* **ROUND**: Ondalık değerleri yuvarlar.
* **SIGN**: Herhangi bir sayının işaretini bize verir. N, P, 0.
* **SIN**
* **SQRT**: Karekök alır.
* **SQUARE**: Karesini alır.
* **TAN**

# 6.2_ABS/SIGN Fonksiyonları

* **ABS**: Mutlak değer fonksiyonudur. Bir sayının mutlak değerini getirir.
* **SIGN**: İşaret fonksiyonudur. Bir sayının işaretini getirir.

## ABS

### Kullanım

#### Input

```
SELECT ABS(-25), ABS(-30.12), ABS(11)
```

#### Output

```
25 | 30.12 | 11
```

### Not:

* Bir sistemde artış ya da azalışları işaretinden bağımsız olarak görmek istiyorsan mutlak değerlerini toplarsın.
* İşaretiyle ilgilenmezsin sadece farkları toplarsın.

## SIGN

### Kullanım

#### Input

```
SELECT SIGN(25), SIGN(-25), SIGN(-25.56), SIGN(0)
```

#### Output

```
1 | -1 | -1.00 | 0
```

# 6.3_ROUND/FLOOR/CEILING Fonksiyonları

* Üç fonksiyon da ondalıklı sayıda virgülden ayrılmayı amaçlar.
* **ROUND**, aşağı ya da yukarı yuvarlama yapar, virgülden sonra kaç karakter alacağını belirler.
* **FLOOR**, aşağı yuvarlar, int değer elde edilir.
* **CEILING**, yukarı yuvarlar, int değer elde edilir.

### Kullanım

```
SELECT ROUND(156.4678, 2), FLOOR(156.4678), CEILING(156.4678)
```

```
>> 156.4700 | 156 | 157
```

### Not:

* 156.4700 çıktısının sebebi **DECIMAL** olarak algılamasıdır. Eğer değeri **FLOAT**a çevirirsen 156.47 olarak çıkartır.

```
SELECT ROUND(CONVERT(FLOAT, 156.4678), 2)
```

```
>> 156.47
```


----

# 6.4_RAND Fonksiyonu

* Random'ın kısaltmasıdır. RAND fonksiyonu da her çalıştırıldığında 0 ile 1 arasında float bir değer üretir.
* Bir rastgeleliğe ihtiyaç duyulan her yerde kullanılır.

### Kullanım

```
SELECT RAND()
```

```
>> 0.901234298... 
```

### Örnek

1-100 arasında rastgele bir int değer üret.

```
SELECT CONVERT(INT, RAND() * 100)
```

```
>> 95 (rastgele)
```

50-100 arasında rastgele bir int değer üret.

```
SELECT 50 + CONVERT(INT, RAND() * 50)
```

```
>> 55 (rastgele)
```
