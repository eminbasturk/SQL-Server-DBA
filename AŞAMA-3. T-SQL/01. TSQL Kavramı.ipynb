* SQL bütün veritabanlarının ortak dilidir.

### Structured Query Language
* Yapısal Sorgulama Dili
* Bütün veritabanlarının ortak dili. 
* 1970'de IBM mühendisleri tarafından geliştirildi.
* Veriler büyüdükçe verilerin programlanması gerekti.

#### Diğer programlama dilleri veri üzerinde çalışamaz mı?
* Çok büyük verilerde verinin veritabanından çekilip üzerinde işlemler yapılması zordur. Veri yerinde işlenmelidir.
* 1 milyar satır veri varsa ve her satır ortalama 1 KB ise 1 milyar x 1 KB = 1 TB veri büyüklüğü demektir.
* KVKK gibi sebepler de veriyi içeride işlemeye zorlamaktadır.
* Veriyi işlemek için farklı veri tabanı yönetim sistemlerinin farklı programlama dilleri vardır.
    * Oracle'da PL/SQL
    * SQL Server'da T-SQL
    * PostgreSQL ve MySQL'de özelleştirilmiş programlama dilleri vardır.

### T-SQL
* T-SQL, SQL Server'ın SQL dilinden evrimleştirilerek oluşturmuş olduğu bir veritabanı programlama dilidir.
* T-SQL ile bir programlama dili ile yapılabilecek hemen hemen her şey yapılabilir.
    * Değişken tanımlama
    * Değişkene değer atama
    * Döngü
    * Cursor
    * Dosya okuma/yazma
    * If blokları
    * Web servislere bağlanma
    * Json verisini okuma/yazma
    * XML verisini okuma/yazma
    * Başka bir veritabanına bağlanıp sorgu çekme
    * CLR ile C# üzerinde yazılan dll'leri kendi içerisinde doğrudan kullanabilme

# 1.9_Örnek Uygulamalar, Fulltext

### FullText Arama
* Veritabanı yönetim sistemleri dikeyde arama yapmak için tasarlanmıştır.
    * Yani arama satırlar üzerinde yapılır.
* Eğer veritabanını doğru şekilde optimize etmişsen bu satırlar üzerinde yaptığın aramalar oldukça hızlı şekilde çalışır.
* Ancak her zaman aramayı dikeyde yapmak istemeyebilirsin. Bazen de yatayda yapmak istersin.
* Örneğin bir kütüphane sisteminde veri tabanında kitaplar var ve kitap içerisinde bir kelime aramak istiyorsun.
    * Her satırda bir kitabın pdf içeriğinin text olarak bulunduğunu düşün.
    * Bir kelime arattığında her bir satırdaki bir kitabın içeriğinde kelime aratman gerekiyor. Yani yatayda bir arama söz konusu.
    * Yataydan kasıt; yanyana kelimeler bir tren vagonu gibi dizilmiş bir şekilde bulunuyor. Yüzbinlerce kelime içerisinde arama patterni aranmak istiyor buna yatayda arama denir. Bunun için standart komutlar işe yaramaz.
    * Mesela **LIKE** fonksiyonu ile aratabilirsin ama çok büyük performans sorunlarına yol açar.
    * Bu yüzden SQL Server ile gelen Fulltext Arama kullanılmalıdır.
