# TSQL Komutları
* SQL haricinde herhangi bir programlama dilinde yapabildiğimiz hemen hemen her şeyi yaptığımız dil **T-SQL**dir.

# 1. DDL Kodları
Data Definition Language - Veri Tanımlama Dili

* Sadece TSQLe özel değildir. SQL dilinin de komutlarıdır. Ancak temel için önemlidir.
* Her SQL Server objesi (database, tablo, kullanıcı, view, procedure) **CREATE** komutu ile oluşturulur, **ALTER** komutuyla değiştirilir ve **DROP** komutuyla silinir. Bu komutlara **DDL komutları** denir.

<img src = "https://i.hizliresim.com/s7ivmru.png" width = 700>

* DDL komutları **KOMUT NESNETÜRÜ NESNEADI** şeklinde tanımlanır. 
* Eğer beraberinde parametre tanımlayacaksan parantez içine yazılır.
    * Tanımlanan nesne tablo ise parantez içine **SÜTUNADI VERİTÜRÜ** şeklinde yazılır. Eğer otomatik artan bir sütun olacaksa **IDENTITY(1,1)** şeklinde tanımlanır.

* Tablo silineceği zaman da **DROP TABLE ISIMLER** şeklinde silinir.
* Sorgu yapmadan önce hangi veritabanını seçeceğini **USE DATABASEADI** ile belirtebilirsin.

<img src = "https://i.hizliresim.com/kt7rf8z.png" width = 700>

### Özet
* SQL Serverda değişken **DECLARE** komutu ile tanımlanır.
* Değişkene bir isim verilir ve başına **@** getirilir.
* Daha sonra **AS** diyerek değişkenin türü tanımlanır.
* **SET** ile değişkenin ismi kullanılarak değişkene değer verilir.
* Tanımlanan bu değişken üzerinden matematiksel, tarihsel ve string işlemleri türüne göre yapılabilir.
* Veritabanından dönen değeri değişkene atarken **SELECT** ile değişken adı ve sütun adı eşlenir. Ancak bunun için tablodan tek satır dönmesi gerekir. Birden fazla değer dönerse son değeri alır.
