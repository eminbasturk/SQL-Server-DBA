# 1. UDF (Kullanıcı Tanımlı Fonksiyonlar) Giriş

* Fonksiyonlar içerisine parametre alır ya da almaz bir sonuç döndürür.
* Sonuç döndürmesi bir SQL sorgusunda satır satır çalışabileceği anlamına gelir.


----
### Basit bir fonksiyon

#### Fonksiyonun tanımlanması

* İki sayıyı toplayıp sonucu döndüren fonksiyonu tanımlama.
* Her SQL Server nesnesi gibi fonksiyonlar da **CREATE** ile tanımlanır. Ancak diğerlerinden farklı olarak fonksiyonun şeması da belirtilir. Burada ilgili şema **DBO**dur.

<img src = "https://i.hizliresim.com/7yu7mlw.png">

* Burada **TOPLA** fonksiyonun ismidir. @SAYI1 ve @SAYI2 değerleri fonksiyonun tanımlanan parametrelidir.
* **RETURNS** fonksiyonun döndüreceği verinin türünü belirtir.
* **BEGIN** ve **END** arasına fonksiyondaki işlemler yazılır ve sonuç **RETURN** ile döndürülür.
* Bu tür tek bir değer döndüren fonksiyonlar **Programmability -> Functions -> Scaler-valued Functions** altında bulunur.

* Diğer bütün fonksiyonlara benzer şekilde kullanılır. Tanımda olduğu gibi başına şema ismi de getirilir.

# 4. İki Parametreli Fonksiyon

### Fonksiyonların Avantajları
* Fonksiyonlar kodu derli toplu hale getirir.
* Fonksiyonlar bir kez tanımlandıktan sonra defalarca kullanılabilir ve isteğe göre değiştirilebilir.

### Fonksiyonların Dezavantajları
* Satır satır çalıştığı için sistemi biraz daha fazla yorar. Ancak doğru indexlerle bunun bir nebze önüne geçilebilir.

----

* Birden fazla hemen hemen aynı işi yapan fonksiyonlar kullanıldığında bu fonksiyonları tek bir fonksiyonda birleşmek mantıklıdır. Aksi takdirde çok karışıklık olur.
* Fonksiyonları farklı amaçlarda kullanmak için örneğin ilk parametreyi değeri alarak ikinci parametreyi de ne yapılacağını belirterek gösterebilirsin.

<img src = "https://i.hizliresim.com/c21gtxz.png">


* Eğer **IF**'den sonra birden fazla satır yazman gerekiyorsa **BEGIN-END** kullanırsın.

* Sonuçta üç fonksiyonun yaptığı işlemleri tek bir fonksiyon kendi başına halledebilir.

<img src = "https://i.hizliresim.com/hhpbtix.png">

# 5. Table Valued Functions

* Evet fonksiyonlar merkezidir, değiştirilmesi kolaydır, kodu sadeleştirir.
* Ancak performans sorunlarına yol açar. Çünkü her bir fonksiyon satır satır çalışır.
* Örneğin üstteki örnekte bulunan 4 fonksiyon her çağrılışında çarpı satır sayısı kadar işlem yapar. Aynı fonksiyonlar tekrar tekrar kullanılsa bile.
* Bunun önüne geçebilmek için bu fonksiyonlar da toplulaştırılabilir. Bu sayede tek bir sütun yani tek bir değer döndürmek yerine komple tablo döndürülebilir. Bu tür fonksiyonlara da zaten **table valued functions** denir.

* **VIEW dışında hiçbir database objesi için Türkçe tanımlamalar kullanma.**

----

* Table valued fonksiyonlar tanımlanırken **RETURNS** olarak *TABLE* verilir ve *BEGIN-END* bloklarına ihtiyaç duyulmadan direkt **RETURN** ile tablo sorgusu döndürülür.

<img src = "https://i.hizliresim.com/5m95qfv.png">

* Bu bir çeşit tablo olduğu için artık sorguda diğer fonksiyonlar gibi kullanılamaz. İçerisine parametre girebildiğin bir tablo gibi davranır.

<img src = "https://i.hizliresim.com/88lukty.png">

* Bu şekilde bir fonksiyonu sorgumuza entegre etmek için ise **CROSS APPLY** denilen *JOIN* yöntemini kullanırız.
* İlgili table valued fonksiyonu tanımladıktan sonra ise kolonlarda fonksiyondaki kolonları aliası ile birlikte **.** ile belirterek çağırabiliriz.

<img src = "https://i.hizliresim.com/fvcqhrg.png">

* <code>SET STATISTICS TIME ON</code> sorgunun ne kadar sürede çalıştığını öğrenmemizi sağlar.
