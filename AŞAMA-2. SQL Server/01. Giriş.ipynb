## Veritabanı Kavramı
* Temel anlamda içerisinde belli bir formatta sorgulanabilir şekilde veri barındıran yazılımlardır.
* Verileri listeler halinde tablo ve satırlarda tutan her yapı aslında kendi çapında veritabanıdır.
* Veritabanı tablolar, sütunlar, satırlar ve indexlerden oluşur.

### Veritabanı Sunucu
* Network üzerinde bir porttan sistemi dinleyen ve gelen komutlara göre istenilen veriyi gönderen bir yazılımdır.
* SQL Server, MySQL, PostgreSQL, Oracle gibi sistemler veritabanı sunucusudur.
* Access, Excel gibi yapılar ise bir sunucu değildir, veritabanıdır.
* Verileri çekmek, değiştirmek ve işlemek amacıyla veritabanı sunuculara komutlarla istek göndeririz.
* **Veritabanı Yönetim Sistemi**, verilerle ilgili her şey için kullanılır verileri yönetmek bu manada doğrudur.

# Microsoft SQL Server
* Microsoftun veritabanı yönetim sistemidir.
* İçerisinde veriler barındırdığı için bir veritabanıdır. 
* İçerisinde bulunan veriler yönetilebildiği için hem bir veritabanı sunucu hem de bir veritabanı yönetim sistemidir.
* İlişkisel veritabanı sistemlerini destekler.
* OLTP sistemidir. Bir kuruluşun günlük işlemlerinin yönetildiği çoğunlukla ilişkisel veritabanı üzerine kurulu sistemlerdir.

# SQL Server - Hizmetçi Benzetmesi
* DBA ile SQL Server ilişkisi
* Server -> Sunucu demek ama Hizmetçi olarak düşünülecek.

<img src = "https://i.hizliresim.com/h5jhzw3.png" width = 500>

* Microsoft SQL Serverı çıkarmadan önce Microsoft Access isimli bir veritabanı vardı.
* Microsoft Access Server çıkarmak yerine Microsoft SQL Server adında yeni bir veritabanı yönetim sistemi çıkardı.

<img src = "https://i.hizliresim.com/bgtibda.png" width = 700>

* Access denilen yapı bir veritabanı yönetim sistemi değildir. Bir veritabanıdır.
* Access üzerinden veri çekilirken istemci bilgisayar olarak kendi kaynaklarımızı (cpu, disk ve ram) kullanırız.
* Ancak sunucu sistemlerde sistem bizden bir komut bekler ve komut doğrultusunda işlem gerçekleştirir.

<img src = "https://i.hizliresim.com/lkhywye.png" width = 700>

* SQL Serverda bir işlem gerçekleştirmek için SQL Servera onun anlayacağı dilde SQL komutu göndeririz.
* SQL Server komutu alır ve kendi kaynaklarını kullanarak işlemi gerçekleştirir ve bizim anlayacağımız şekilde veriyi istemci bilgisayara gönderir.
* Client-server mimarısı, veritabanı yönetim sistemi yapısı bu şekilde çalışır.
* Veritabanlarının global dili SQL'dir. Yani bir veritabanı yöneteceksen sunucuya derdini en iyi şekilde anlatman gerekiyor. Bunun yolu da SQLi çok iyi bilmekten geçiyor.
* Ancak SQL Serverın ana dili T-SQL'dir. T-SQL ne kadar iyi bilirsen SQL Server veritabanını o kadar iyi yönetirsin.
* SQL Server ile DBA arasında sadece komutu vereyim de kendi kaynaklarını kullanarak bana veriyi getirsin diye monolog bir olay yoktur. Daha çok karşılıklı konuşmaya dayalı bir diyalog vardır.

#### Örnek
* Diyelim SQL Server'dan bi veri istedik.
* Sistem bu veriyi önce RAMde arar bulamazsa Diske gider. Disk yavaştır ve işlemleri gerçekleştirmek için kuyruğa atar ve sırada bekletir. Bundan dolayı sonuç geç gelir.

<img src = "https://i.hizliresim.com/nm4ji3h.png" width = 800>

* Diskler kapasitesi düşük olduğu için ramlerden yavaş çalışır.
* Yetersiz olduğu için herkesin isteği olduğu anda kullanıcıları sıraya atar ve kuyruk oluşturur.
* Bu kuyruk uzunluğuna **ortalama disk kuyruk uzunluğu** (avg disk que length) denir. Bir performans parametresidir.
* AVG DQL < 1 beklenir. Eğer büyükse SQL Server, 'disk performansın yetersiz, bu diski bir şekilde hızlandır' der. Sonucunda daha çok disk konulabilir veya daha hızlı bir ssd disk konulabilir.

<img src = "https://i.hizliresim.com/r75137p.png" width = 800>

* Eğer veri dağınık şekilde duruyorsa da getirilmesi uzun sürer çünkü önce sıralı hale getirilmelidir.
* Verinin sıralı bir şekilde durmasına **index** denir.
* Index bozulması **fragmantasyon** olarak adlandırılır.
* Bu da bir performans parametresidir.

<img src = "https://i.hizliresim.com/b5s3se3.png" width = 800>

* SQL Server RAMde çalışır. RAM yetersiz ise verinin getirilme süresi uzun sürer.
* Veri diskten alınır ve işlemler RAM üzerinde gerçekleşir. RAMi ihtiyacı kadar geniş tutmak gerekir.
* Ram ihtiyacı olduğunun ölçüsü **Page Life Expectancy (PLE)** dir.
* SQL Serverin en küçük yapıtaşı 8 kblik pagelerdir. 
* Okunmak istenen veri rame alındığında ramde pageler bulunur. Başka bir veri alındığı zaman ram küçükse eski pageleri atmak gerekir. Her gelen yeni page yani yeni veriler diğer pageleri atar. 
* Bir pagein ortalama ramde kalma süresi **Page Life Expectancy**dir. Performans parametresidir.
* PLE'nin toplam memory/4*300'den büyük olması beklenir.
    * 64 GB RAMi olan makinenin PLE değeri 4800 GB'den büyük olması gerekir.
    
<img src = "https://i.hizliresim.com/pi1gomp.png" width = 800>

* SQL Server performansına en çok etki eden parametredir. Ram bus hızıdır. Bu değer ne kadar artar ise performans aynı oranda artar.
* Diskten data okunur. Data rame alınır. Ramdeki data CPU ile işlenir.
    * En yavaş disk, işlemek için rame almak gerekiyor. CPU zaten ramden daha hızlı. Dar boğaz noktası ramdir.
    * O yüzden ram ne kadar hızlıysa sistem o kadar hızlıdır.
* Eğer CPU sayısı iki katına çıkarılırsa sorgunun gelme hızı iki katına çıkmaz. 
* Disk kapasitesi iki katına çıkarılırsa sorgunun gelme hızı iki katına çıkmaz.
* RAM, 64 GB'dan 128 GB'a çıkarılırsa sorgunun gelme hızı iki katına çıkmaz.
* Ancak RAM'in bus hızı 2100 MGHz'den 3000 MGHz'e çıkarılırsa neredeyse %50 oranında performans artışı sağlanır.

<img src = "https://i.hizliresim.com/kliibcv.png" width = 800>

* SQL Server matematiksel işlemler yapmamasına veri okumasına rağmen CPU sayısı önemlidir.
* SQL Server özellikle Join işlemlerinde paralel çalışır. CPU sayısı çok olursa paralel işlemler hızlı çalışır.
* Join işlemlerinde her tablo için bir CPU kullanılır.
* Dolayısıyla performans artışı için CPU çekirdeği artırılmaldıır.

<img src = "https://i.hizliresim.com/eefen3g.png" width = 800>

* Daha yeni jenerasyon bir CPU'nun clock speed değeri daha hızlı olacağı için performansta artış sağlanır.

<img src = "https://i.hizliresim.com/fe1oqx4.png" width = 800>

* Yeni datalar geldikçe, eskileri değiştikçe indexler tekrar bozulur ve düzeltilmesi gerekir.

<img src = "https://i.hizliresim.com/cmtjkgo.png" width = 800>

* Indexlerin düzeltilme işlemi Index Rebuild olarak adlandırılır. Bunun için periyodik bir bakım planı oluşturmak gerekir.

<img src = "https://i.hizliresim.com/bn0sml4.png" width = 800>

* SQL cümlelerini teker teker göndermeye **Ad Hoc Query** denir.
* SQL cümlelerini toplu halde göndermeye **Stored Procedure** denir. Performans için çok önemli bir SQL Server nesnesidir.


**SQL Serveri iyi yönetebilmek için onun dilinden iyi anlıyor olmak gerekir.**

* Bir emir alır, o emri gerçekleştirir ve nasıl yaptığını anlatarak önerilerle gelir. Tıkandığı noktaları belirtir ve iyileştirme bekler.

### Veritabanı sorgularını hızlandırmak için;
* Daha çok disk eklenebilir veya daha hızlı bir ssd konulabilir.
* Veriler indexlenebilir ve index bozulmaları kontrol altına alınabilir.
* RAMin kapasitesi arttırılabilir.
* RAMin bus hızı (en çok etki eden) arttırılabilir.
* İşlemcinin çekirdek sayısı arttırılabilir veya üst jenerasyona geçilebilir.
* Periyodik bakımlar yapılabilir.
