# 1. Büyük Veri Uygulamaları

### SQL Server Profiler

* SQL Server içerisinde anlık olarak çalışan SQL ya da T-SQL cümlelerini bize gösteren araçtır.
* **Trace** sistemi izlemek demektir.

<img src = "https://i.hizliresim.com/1zifywa.png">

<img src = "https://i.hizliresim.com/qfvnhf5.png">

#### Sağladığı avantajlar
* Sistem üzerinde belirli şeyler takip edilebilirdir
* A kişisinin yaptığı işlemler takip edilebilir
* B programının yaptığı işlemler takip edilebilir
* Sistemi yoran sorgular takip edilebilir
* Sistem üzerinde uzun süren sorgular tespit edilebilir
* Son kullanıcı ile aramızda köprü olan bir yapıdır

### Örnek-1
* **HR** databaseindeki **WORKERS** tablosunda bulunan çalışanların **WORKER_TRANSACTIONS** tablosundan elde ederek ulaştığın son giriş çıkış hareketini kontrol ettiğini düşün.

<img src = "https://i.hizliresim.com/e95pn8u.png">

* Kullanıcıdan bu sorgunun uzun sürdüğüne dair sorguyu hızlandırmak üzere bir geri bildirim aldın diyelim.
* Kullanıcının çalıştırdığı sorguyu bilmiyoruz. Kullanıcı da hangi sorgunun çalıştığını bilmiyor. Sadece bu yavaş yükleniyor diye bize bir geri bildirim gönderiyor.
* Biz bu kullanıcının yaptığı işlemleri SQL Server Profiler'dan kontrol edebiliriz.

### Çözüm
* Yeni bir trace başlatıp işlemi gerçekleştiren kullanıcının **hostname**ini filtreleyip yaptığı işlemleri görüntüleyebiliriz.

<img src = "https://i.hizliresim.com/1zlk63x.png">

* Kullanıcı dolaylı yoldan sorguyu çalıştırdığında şekildeki gibi sonlanması uzun süren bir sorgu olduğunu görebiliyoruz.
* Bu noktada sorguyu kendi ekranımıza alıp çalıştırdığımızda gerçekten yaklaşık 6 saniye sürdüğünü fark ediyoruz.

<img src = "https://i.hizliresim.com/noly7ca.png">

* Execution plan çıkartarak biz hızlandırmak için herhangi bir önerisini olup olmadığına bakabiliriz.
* Eğer önermiyorsa kendimiz sorgumuzu inceleyerek gerekli indexi oluşturabiliriz.

<img src = "https://i.hizliresim.com/gz2ykww.png">

* Sorgumuzda **WORKERID**'ye göre filtreleme, **DATE_**'e göre de bir sıralama var bu yüzden bu kolonları *index key columns* olarak, **IOTYPE**'ı elde etmek istediğimiz için de bu kolonu *included column* olarak işaretleyerek indeximi oluşturabiliriz.

<img src = "https://i.hizliresim.com/a2mfxiy.png">

* Aynı sorguyu şimdi çalıştırdığımızda ise yaklaşık 1 saniyede sonuca ulaşabiliriz.

<img src = "https://i.hizliresim.com/iyqg93p.png">

* Bu durumda artık kullanıcı sorguyu çalıştırdığında beklemek zorunda kalmadan direkt çıktıyı görebilir.
* SQL Server Profiler sayesinde kullanıcımın yaşadığı sıkıntıyı tespit edip doğrudan iyileştirme gerçekleştirebildim.

### Örnek-2

* Bir de kullanıcıdan bağımsız genel olarak sistemi yoran sorgular veya işlemler neler buna bakmak isteyebilirsin.
* Bunun için de başka bir filtre bulunur. Diyelim ki 1 saniye (1000 ms) ve üstü sorgular benim sistemimi yoruyor. Bu şekilde filtreyeleyebilirsin;

<img src = "https://i.hizliresim.com/502bdcz.png">

* Mesela bu sefer Excel üzerinden çekelim veriyi, yine yaklaşık 6 saniye sürer ve bu sefer trace ekranımıza doğrudan düşer.
* Artık bu ekrandan tek tek hangi kullanıcı 1 saniye veya daha fazla süren sorgu çekmiş takip edebilirim.
* Tekrar index oluşturup sorguyu çalıştırdığımızda index oluşturma işlemimiz buraya düşer ancak sorgumuz 1 saniyeden kısa sürede çalıştığı için buraya düşmez.

### Örnek-3

* Peki sistemi izlemek için ben bu profilerın başında durmak zorunda mıyım?
* Hayır trace oluştururken *Save to file* diyerek bir .trc dosyasına kayıt edip 1 gün sonra *Open Trace File* diyerek dün yapılan işlemleri kontrol edebilirsin.

<img src = "https://i.hizliresim.com/1ih0joh.png">

* Peki bunu analiz edebilmek adına daha kolay çalıştırabilmenin bir yolu var mı?
* Evet, var. *Save to file* yerine *Save to table* diyerek ve veritabanı ve tablo ismi vererek bir tabloya kayıt edebilirsin.
* Ardından bu trace üzerinde işlemler yapıp tracei durdurduktan sonra oluşan tablodan sorgular gerçekleştirip gerekli aksiyonları alabilirsin.

<img src = "https://i.hizliresim.com/4f0xs6z.png">

# 2. Türkiye Eczane Örneği

### SQL Server'da Büyük Veri İşlemleri

* Anadolu Ajansı'nın yaptığı bir habere göre 2016 yılında Türkiye'de 2 milyar kutu ilaç satıldı.

<img src = "https://i.hizliresim.com/qdps9ls.png">

* Önceden ilaçların üzerinde her ilaç için aynı olan çizgi barkodlar bulunurdu. Günümüzde ise ilaçlar aynı olsa bile her birinde farklı karekodlar bulunuyor. Bu değişimin amacı yasadışı ilaç satımının önüne geçmek ve ilaçları takip etmek. 

<img src = "https://i.hizliresim.com/qiq9okf.png">

* İlaçların takibi için avantajlı olan bu sistemde de her bir ilacın karekodu ayrı olduğu için ciddi bir maaliyet sorunu var.
* Türkiye'de 2017'de satılan ilaç sayısı yaklaşık 2 milyar.
* Türkiye'de satılan ilaç çeşidi sayısı yaklaşık 11 bin.
* 2 Milyar / 365 = 5.479.000 adet (günlük satılan)
* 5.479.000 / 24 = 228.000 adet (saatlik satılan)
* 228.000 / 60 = 3.815 adet (dakikada satılan)
* 3815 / 60 = 65 adet (saniyede satılan)
* Yani Türkiye'de saniyede ortalama 65 kez barkod okutuluyor. 

# 3. İlaç Veritabanı Oluşturma

* Bir ilaç barkodu okutulduğunda sistem sağlık bakanlığının veritabanı sunucusuna bağlanıp o bilgiyi çeker.
* 1 yılda 2 milyar ilaç satıldığını düşünürsek minimum 2 milyar satırlık tablodan bu sorgu gerçekleştirilir.

### Simülasyon

* 2 milyarlık veri yerine 100 milyonluk veri kullanılacak ne olur ne olmaz bilgisayarımızın başına bir şey gelmesin. (**MEDICAL** veritabanı)
* **DRUGS**, ilaçları barındıran tablo. 
    * İlaçların rastgele isimleri ve rastgele fiyatları tutuluyor.
    * 4776 adet ilaç var.
* **DRUGBARCODE**, ilaçların barkodlarını barındıran tablo.
    * İlaçların idsini ve guid formatında barkodları
    * 100 milyon satır barındırır.
        * <code>COUNT(*)</code> ile de bakabilirsin.
        * <code>SP_SPACEUSED 'DRUGBARCODE'</code> ile de daha hızlı sürede bakabilirsin.
* **DRUGBARCODE** içerisinden bir barcode çektiğinde (ilaç okuttuğunda) senin bilgisayarında yaklaşık 23 saniye sürüyor. Eczanelerin 24 saat çalıştığını varsayıp saniyede ortalama 65 barcode okutulduğunu düşündüğümüzde bu süre beklenilecek bir süre değildir.
* **BARCODES** tablosunda 100 milyonluk barcode verisinden 1000 tane barcode bulunuyor.
    * Simülasyon uygulaması bu tablodan rastgele barcode çekecek ve barcodeun bilgisini okutup bize sonuç getirecek.

# 4. Sanal Eczane Uygulaması

* Uygulamada istediğin sayıda sanal eczane oluşturarak barkodları okutmaya başlayabilirsin.
* Az sayıda sanal eczane varken nispeten düşük sürelerde okutulan barkodlar, eczane sayısı arttıkça sıraya girileceği için okutma süreleri artar.
* Elimizde büyük bir veri var. Bu veriyi çoklu şekilde kullanan bir yapı var. Bundan kaynaklanan da bir performans sıkıntısı var. Database admin olarak bu sorun bize geldi diyelim ne yaparız?

### Çözüm

* Bir yerde performans problemi varsa onun hangi sorgudan kaynaklandığını görmek için sql server profiler kullanabiliriz.

* SQL sorgusunu profilerdan alıp bi inceleyelim.
