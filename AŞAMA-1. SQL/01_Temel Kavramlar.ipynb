# Temel Kavramlar

## SQL
**S**tructured **Q**uery **L**anguage (Yapısal Sorgulama Dili)
* SQL dili verinin alfabesidir.
* Elimizdeki veri seti dağınık değil belirli bir yapıya sahip yani **yapısal**, veri analiz edilmek isteniyor bunun adı **sorgulama**, bu işi yaparken de belirli kuralları ve syntaxı olan bir **dil** kullanılacak.

## SQL nedir?
* Bir veritabanındaki verileri sorguladığımız bir çeşit kodlama dilidir.

## Veritabanı nedir?
* İçerisinde belirli bir formatta sorgulanabilir şekilde veri barındıran yazılımlardır. (Excel, Access, Notebook gibi)
* Verileri listeler halinde tablo ve satırlarda tutan her yapı aslında kendi çapında veritabanıdır.

<img src = "https://i.hizliresim.com/4r7dj5f.png" width = 600>

## Veritabanı nelerden oluşur?
* Tablolar
* Sütunlar
* Satırlar
* Indexler

## Veritabanı sunucu nedir?
* Bir yazılımdır.
* İçinde bulunan verileri sorgulamak istediğimizde bize bir sonuç döndürür.
* Network üzerinde bir porttan sistemi dinler ve gelen komutlara göre istenilen veriyi gönderir.
* SQL Server, MYSQL, PostgreSQL, Oracle gibi sistemler veritabanı sunucusudur.
* Access, Excel gibi yapılar ise bir sunucu değil sadece bir veritabanıdır.
* Her veritabanı sunucu aslında bir veritabanıdır.

### İstemci Bilgisayar - Veritabanı Sunucu
* Veritabanı sunucu, istemciden gelen komutlara göre cevap verebilen bir yapıya sahiptir.
* Veritabanı sunucu kendisine bir istemcinin bağlanmasını bekler.
* İstemci, sunucuya bağlanmak istediğini belirterek bir kullanıcı adı ve şifre bilgisi gönderir.
* Sunucu kendi içinde bu kullanıcı adı ve şifreye sahip bir yetkili arar eğer bulamazsa hata döndürür, eğer bulursa bağlantıyı kabul eder.

<img src = "https://i.hizliresim.com/m7nocap.png" width = 600>


* Bu bağlantı kabulünden sonra istemci ve sunucu kendi arasında birbirinin dilinden anlaşabileceği şekilde konuşmaya başlar.
* Bu konuşma SQL komutları ile gerçekleşir.

#### Örnek
* Veritabanı sunucusunda **MUSTERILER** adlı tablo bulunsun.
* İstemci bu tabloyu görüntülemek için bir SQL komutu yazarak tabloyu görmek ister.
* Sunucu da istemcinin bu isteğini anlayarak veritabanından bu veriyi çekerek istemcinin isteğini yerine getirir.
* Sunucu bu isteği istemcinin anlayacağı bir formatta yerine getirir.
* İstemci de bu formatlı veriyi derleyerek kullanıcının anlayabileceği şekilde bir liste halinde ekrana yansıtır.

<img src = "https://i.hizliresim.com/rac1t4b.png" width = 600>

* Yukarıda yazan **SELECT** komutu MUSTERILER tablosunda ne var ne yok bana getir anlamı taşır.
* Temel anlamda bütün veri tabanlarında yapılan işlem aynıdır.
    * Veriyi çekme (**SELECT**)
    * Veriyi silme (**DELETE**)
    * Veri ekleme (**INSERT**)
    * Veriyi güncelleme (**UPDATE**)
    
#### Örnek
* Veritabanına bir veri eklenmek istendiğinde bu işlem aşağıdaki şekilde gerçekleşir.

# İlişkisel Veritabanı (RDMS)
**Relational Database Management System**
* Veriler veri tabanlarında tablolar halinde tutulur.
* Tablolar birbirleriyle ilişki içinde bulunabilir. İlişkisel veritabanı bu tablolar arasındaki ilişkiyi tarif eder.

<img src = "https://i.hizliresim.com/4ssav1f.jpg" width = 300>

* Tekrar eden verileri tekilleştirmek amacı ile yapılandırılan veritabanı sistemleridir.
* Eğer veriler tekrar edilirse; zaman kaybı yaşanır, bellekte yer kaplar, hata yapılma ihtimali artar.

### Örnek
* Bir mağaza düşün.
* Mağazaya kayıtlı müşteriler var ve o müşteriler zaman zaman mağazadan alışveriş yapıyor.
* Yeni bir sipariş oluşturulacağı zaman müşteriler tablosundan müşteri çekiliyor. Teslimat şekli, sipariş tarihi ve ödeme yöntemi belirlendikten sonra ürünler eklenmeye başlanıyor.
* Ürünler ürünler tablosundan ekleniyor ve fiyatlara göre fiş oluşturulmaya başlanıyor.

<img src = "https://i.hizliresim.com/5o8fgcm.jpg" width = 600>

* Yukarıdaki görüntü son kullanıcının gördüğü görüntüdür.
* Aşağıdaki ise işlemlerin veritabanındaki karşılığı bulunur.

<img src = "https://i.hizliresim.com/f54h30p.jpg" width = 600>

* Yukarıdaki tabloda sipariş bilgileri tutulur.
* Bu tablo manuel olarak önceki görseldeki gibi oluşturulur.
    * Fişin id'si, 
    * fiş numarası, 
    * ödeme türü (1) -kredi kartı yazmak yerine 1 yazılır, bellekte az yer kaplar ve yanlış yazımlara açık değildir- 
        * ödeme türleri tablosundan gelir, 
    * müşteri id'si var -müşteri ismi falan tutmak gereksiz-,
        * müşteriler tablosundan gelir
    * teslimat türü var (1), 
        * teslimat türleri tablosundan gelir
    * tarih elle girilir, 
    * net fiyat siparişin net fiyatını verir, 
        * detaylardan gelir toplanarak eklenir
    * toplam kdv tutarı, 
        * detaylardan gelir toplanarak eklenir
    * toplam fiyat 
        * detaylardan gelir toplanarak eklenir

<img src = "https://i.hizliresim.com/p79elrh.png" width = 600>

* Yukarıdaki tabloda sipariş detayları tutulur.
    * Detaylarda her bir ürünün id'si bulunuyor -eklenen-, 
    * fiş tablosu ile bağlantı fişin id'si ile sağlanmış, 
    * ürünler tablosundan çekildiği için ürün id var, 
    * ürün adeti giriliyor, 
    * ürün fiyatı
        * ürünler tablosundan
    * kdv oranı
        * ürünler tablosundan
    * kdv miktarı, ürün fiyatı ile kdv oranından hesaplanan
    * net toplam, ürün adedi ile ürün fiyatı çarpımı
    * toplam fiyat
    
* **PRIMARY KEY** bir tablodaki benzersiz sütunu temsil eder.
* **FOREIGN KEY** başka tablodaki benzersiz sütunu temsil eden sütundur.
* Bu şekilde oluşan en az iki tablolu yapıya **master-detay yapı** denir.
* **ORDERFICHE** tablosu master tablo, **ORDERLINE** tablosu detay tablodur.

<img src = "https://i.hizliresim.com/14gnxo1.jpg" width = 600>

* Tüm tablolar yukarıda gösterilmektedir.
* Bir müşterinin birden fazla adresi olabileceği için **CUSTOMER** tablosu ile **ADDRESS** tablosu arasında **1-n**'lik bir ilişki söz konusudur.
* Bir sipariş fişinin altında birden fazla sipariş detayı olabileceği için **ORDERFICHE** tablosu ile **ORDERLINE** tablosu arasında **1-n**'lik bir ilişki söz konusudur.
* Bir ürün tek bir satırda sipariş verilebileceği için **ITEM** tablosu ile **ORDERLINE** tablosu arasında **1-1**'lik bir ilişki söz konusudur.
* Bağlantılar aşağıdaki gibi gözükür.

<img src = "https://i.hizliresim.com/me652gg.jpg" width = 600>

* İlişkisel Veritabanı budur.
* Tablolar arasında tekrar etmeyi önlemek amacıyla ilişkiler bulunur.

# Temel Kavramlar Bölüm Sonu Soruları

#### Veritabanı nedir?
Verilerin belli bir yapıda tutulduğu sorgulamaya açık yazılımlardır.

#### SQL Kavramı nedir?
Yapısal sorgulama dili anlamına gelir. Verinin ana dilidir.

#### Veritabanı Yönetim Sistemi nedir?
Veritabanlarını SQL veya sistemlerin kendi diliyle sorgulayıp veritabanlarını yönettiğimiz sistemlerdir.

#### Veritabanı Sunucusu nedir?
Veritabanlarını içinde bulundurup bir dizi izinden geçtikten sonra veritabanlarının sorgulanmasına müsade eden yapılardır.

#### Client-Server Mimarisi nasıl çalışır?
Client izin ister, Server izin verir. Client sorgular, Server yanıtlar.

#### İlişkisel Veritabanı nedir?
Veri tekrarlarının önüne geçmek amacıyla veritabanındaki tabloların ilişkiler içinde bulunduğu yapıdır.

#### Veritabanı Yönetim Sistemi yazılımları hangileridir?
MSSQL, MySQL, PostgreSQL, SQLite, Oracle gibi.
