**Sana ait olmayan tüm ülkeleri güncellemek için bir transaction oluştur, ardından rollback yap**

**CountryName** kolonunu evin olarak adlandırdığın ülkenin yanındaki (**My Country**) kelimelerini içerecek şekilde ayarlamak için **tblCountry** tablosunda bir **UPDATE** ifadesi kullan.

<img src = "https://i.hizliresim.com/lf30egb.png">

Şimdi bir transactiondaki diğer ülkelere (**Holiday destination**) eklemek için kodu değiştir. Update ifadesini çalıştırdıktan sonra:

* 1 + 1 = 2 ise transaction'ı rollback et; veya
* diğer türlü commit et

Bu sorguyu çalıştırdığında hiçbir şey değişmemelidir: 

<img src = "https://i.hizliresim.com/b7swklz.png">

* Ülkeni hariç tutarken, değiştirilen adı hesaba katmak için joker karakterler kullanmayı unutma!

1 + 1 = 3 olup olmadığını test etmek için koşulu değiştir ve yeniden çalıştır. Fazladan metnin eklenip eklenmediğini kontrol et: 

<img src = "https://i.hizliresim.com/ltr01ms.png">
