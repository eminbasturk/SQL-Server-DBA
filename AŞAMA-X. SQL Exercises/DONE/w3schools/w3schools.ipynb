# > SQL Join

## Exercise 1:
**Eksik parçaları, iki tablo arasındaki ilişki olarak her iki tablodaki CustomerID alanını kullanarak Orders ve Customers tablosunu birleştirmek için JOIN cümlesine ekle.**

<code>SELECT * 
 FROM Orders
 LEFT JOIN Customers
 ON Orders.CustomerID = Customers.CustomerID</code>

## Exercise 2:
**Her iki tabloda da eşleşme olan iki tablodan tüm kayıtları seçmek için doğru JOIN kalıbını kullan.**

<code>SELECT * 
 FROM Orders
 INNER JOIN Customers
 ON Orders.CustomerID = Customers.CustomerID</code>

## Exercise 3:
**Customers tablosundaki tüm kayıtları ve Orders tablosundaki tüm eşleşmeleri seçmek için doğru JOIN kalıbını kullan.**

<code>SELECT * 
 FROM Orders
 RIGHT JOIN Customers
 ON Orders.CustomerID = Customers.CustomerID</code>
