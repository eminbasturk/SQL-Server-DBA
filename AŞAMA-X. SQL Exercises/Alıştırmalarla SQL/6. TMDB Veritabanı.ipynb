#### COMPANY
* Yapımcı şirketlerin tutulduğu tablo
    * *ID* : Şirketin dbde tutulduğu id
    * *DESCRIPTION_* : Şirketin tanımlaması
    * *HEADQUARTERS* : Genel merkezi
    * *HOMEPAGE* : Web sayfası
    * *COMPANYID* : TMDB'de şirketin idsi
    * *LOGO_PATH* : TMDB'de şirketin logo dizini
    * *NAME_* : Şirketin adı
    * *ORIGIN_COUNTRY* : Şirketin bulunduğu ülke
    * *PARENT_COMPANY* : Eğer varsa şirketin üst şirketi
    
#### COUNTRY
* Ülkelerin tutulduğu tablo (şirketlerin ve aktörlerin)
    * *ID* : Dbde tutulduğu id
    * *CODE* : Ülke kodu
    * *COUNTRY* : Ülke ismi
    
#### DEPARTMENT
* Set çalışanlarının mesleklerinin departmanları
    * *ID* : Dbde tutulduğu id
    * *DEPARTMENT* : Departman ismi
    
#### COLLECTION
* Serisi olan filmlerin tutulduğu seri tablosu
    * *ID* : Dbdeki id
    * *COLLECTIONID* : TMDBde tutulduğu id
    * *NAME_* : Seri ismi
    
#### GENRE
* Film kategorilerinin tutulduğu alan
    * *ID* : Dbde tutulduğu id
    * *GENREID* : Kategorinin idsi
    * *GENRE* : Kategori ismi
    
#### MOVIE_COMPANY
* Bir filmin birden fazla yapımcı şirketi olabileceği için bunu ayrıştırmak adına üretilen tablo
    * *ID* : Dbdeki id
    * *FILMID* : Filmin idsi
    * *COMPANYID* : Şirketin idsi
    
#### MOVIE_COUNTRY
* Filmin çekildiği ülkeleri tutan tablo
    * *ID* : Dbdeki id
    * *FILMID* : Filmin idsi    
    * *COUNTRYCODE* : Ülke kodu
    
#### JOB
* Set görevlisinin mesleğinin tutulduğu yer
    * *ID* : Dbde tutulduğu id
    * *DEPARTMENT* : Mesleğin departmanı
    * *DEPARTMENTID* : Mesleğin departmanı idsi   
    * *JOB* : Meslek
    
#### KEYWORD
* Filmlerde kullanılan anahtar kelimeler
    * *ID* : Dbde tutulduğu id
    * *KEYWORD* : Anahtar kelime
    
#### LANGUAGE_
* Filmlerde kullanılan dillerin tutulduğu tablo
    * *ID* : Dbde tutulduğu id
    * *CODE* : Dil kodu
    * *NAME* : Dil
    
#### MOVIE
* Filmlerin tutulduğu tablo
    * *ID* : Filmin dbde tutulduğu id
    * *FILMID* : Filmin TMDB'deki idsi
    * *TITLE* : Filmin başlığı
    * *ADULT* : Filmin yetişkin olup olmaması
    * *BACKDROP_PATH* : Filmin yatay posterinin urlsi
    * *BUDGET* : Filmin bütçesi
    * *HOMEPAGE* : Filmin web sitesi
    * *IMDB_ID* : Filmin IMDBdeki idsi
    * *ORIGINAL_LANGUAGE* : Filmin orijinal dili
    * *ORIGINAL_TITLE* : Filmin orijinal adı
    * *OVERVIEW* : Filmin kısa özeti
    * *POPULARITY* : Filmin popülerlik değeri
    * *POSTER_PATH* : Filmin dikey posterinin urlsi
    * *RELEASE_DATE* : Vizyona girme tarihi
    * *REVENUE* : Filmin hasılatı
    * *RUNTIME* : Dakika olarak filmin süresi
    * *VOTE_AVERAGE* : Filmin ortalama puanı
    * *VOTE_COUNT* : Film için yapılan oylama sayısı
    * *STATUS_* : Filmin durumu
    * *TAGLINE* : Filmle alakalı slogan
    * *COLLECTIONID* : Filmin serisi varsa seri idsi tutulur
    
#### MOVIE_GENRE
* Bir filmin birden fazla kategorisi olabilir
    * *ID* : Dbdeki id
    * *FILMID* : Filmin idsi
    * *GENREID* : Kategorinin idsi
    
#### MOVIE_KEYWORD
* Film anahtar kelimelerinin tutulduğu alan
    * *ID* : Dbde tutulduğu id
    * *FILMID* : Filmin idsi
    * *KEYWORD* : Anahtar kelime
    
#### MOVIE_RELEASEDATE
* Filmin belirli zamanlarda gösterime girilmesinin tutulduğu tablo
    * *ID* : Dbdeki id
    * *FILMID* : Filmin idsi
    * *CERTIFICATION* : Sertifika
    * *RELEASE_DATE* : Gösterim tarihi
    * *TYPE* : 1. Premiere, 2. Theatrical (limited), 3. Theatrical, 4. Digital, 5. Physical, 6. TV
    * *LANGUAGE_* : Gösterim dili
    * *NOTE* : Gösterime girdiği yer
    
#### MOVIE_COUNTRY
* Filmin çekildiği ülkeleri tutan tablo
    * *ID* : Dbdeki id
    * *FILMID* : Filmin idsi    
    * *COUNTRYCODE* : Ülke kodu    

#### MOVIE_REVIEW
* Filmlerle alakalı yapılan yorumların tutulduğu tablo
    * *ID* : Dbdeki id
    * *FILMID* : Filmin idsi
    * *AUTHOR* : Yorumu yapan kişi
    * *COMMENTID* : Yorumun idsi
    * *URL_* : Yorumun linki
    * *CONTENT_* : Yorum
    
#### MOVIE_SPOKEN_LANGUAGES
* Filmin içinde konuşulan dillerin tutulduğu tablo
    * *ID* : Dbdeki id
    * *FILMID* : Filmin idsi
    * *SPOKEN_LANGUAGE_ID* : Konuşulan dilin idsi
    
#### PERSON
* Aktör ve çalışanların tutulduğu tablo
    * *ID* : Dbdeki id
    * *PERSONID* : TMDBdeki id
    * *NAME_* : İsmi
    * *DEPARTMENTID* : Departman id
    * *DEPARTMENT* : Departman
    * *BIRTHDAY* : Doğum tarihi
    * *ALSO_KNOWN_AS* : Şu isimle de bilinir
    * *DEATHDAY* : Hayatını kaybetmişse ölüm tarihi
    * *GENDER* : Cinsiyet (1 kadın, 2 erkek)
    * *BIOGRAPHY* : Biyografisi
    * *PROFILE_PATH* : Resminin bulunduğu url
    * *POPULARITY* : Popülerliği
    * *PLACE_OF_BIRTH* : Doğum yeri
    * *ADULT* : Yetişkin filmlerinde mi oynuyor
    * *IMDB_ID* : IMDB idsi
    * *HOMEPAGE* : Web sayfasının urlsi
    * *COUNTRYCODE* : Ülke kodu
    * *COUNTRY* : Ülke
    
#### CAST_CREDIT
* Filmde rol alan aktörlerin tutulduğu tablo
    * *ID* : Dbdeki id
    * *PERSONID* : Kişinin idsi
    * *CREDITID* : 
    * *CASTID* : 
    * *CHARACTER_* : Karakterin ismi
    * *GENDER* : Cinsiyeti
    * *NAME_* : Aktörün adı
    * *ORDER_* : Başrol sırası
    * *FILMID* : Filmin idsi
    
#### CREW_CREDIT
* Filmde görev alan personel listesi
    * *ID* : Dbdeki id
    * *FILMID* : Filmin idsi    
    * *CREDITID* : 
    * *DEPARTMENTID* : Departman idsi
    * *PERSONID* : Kişinin idsi
    * *JOBID* : İşinin idsi
    * *NAME* : Kişi
    
#### MOVIE_RECOMMEND
* Filme alternatif filmler
    * *ID* : Dbdeki id
    * *FILMID* : Filmin idsi
    * *RECOMENDED_FILMID* : Önerilen filmin idsi
    * *RECOMENDED_TITLE* : Önerilen film
    * *RECOMENDED_ORIGINAL_TITLE* : Önerilen filmin orijinal ismi
    * *RECOMENDED_ORIGINAL_LANGUAGE* : Önerilen filmin orijinal dili
    
#### MOVIE_SIMILAR
* Filme benzer filmler
    * *ID* : Dbdeki id
    * *FILMID* : Filmin idsi
    * *SIMILAR_FILMID* : Benzer filmin idsi
    * *SIMILAR_TITLE* : Benzer filmin ismi
    * *SIMILAR_ORIGINAL_TITLE* : Benzer filmin orijinal ismi
    * *SIMILAR_ORIGINAL_LANGUAGE* : Benzer filmin orijinal dili
